# 1. Missing Values

```python
df = dataset.copy()
cols, per = [],[]
for col in df.columns:
    
    if(df[col].isnull().sum() > 0):
#         print(col, 100*(df[col].isnull().sum()/len(df)).round(2),"%")
        cols.append(col)
        per.append(100*(df[col].isnull().sum()/len(df)).round(2))
        
        
plt.barh(range(len(cols)),per, tick_label = cols)
plt.title('Percentage of null values')
plt.show()
```

```python
#First lets handle categorical nan values
cols, per = [],[]
features = df.select_dtypes(include ='object').columns
features_nan = [c for c in features if df[c].isnull().sum() > 1]
for col in features_nan:
    
#         print(col, 100*(df[col].isnull().sum()/len(df)).round(2),"%")
    cols.append(col)
    per.append(100*(df[col].isnull().sum()/len(df)).round(2))
        
        
plt.barh(range(len(cols)),per, tick_label = cols)
plt.title('Percentage of null values')
plt.show()
```

```python
#Replace missing categorical values with a new label
def replace_cat_feature(df, feature):
    data = df.copy()
    data[feature]= data[feature].fillna('Missing')
    return data
df = replace_cat_feature(df, features_nan)
df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
## Replacing numerical missing values
cols, per = [],[]
num_features_with_nan = [f for f in df.select_dtypes(exclude = 'object') if df[f].isnull().sum() > 1]
for col in num_features_with_nan:
    
    print(col, 100*(df[col].isnull().sum()/len(df)).round(4),"%")
    cols.append(col)
    per.append(100*(df[col].isnull().sum()/len(df)).round(4))
        
        
plt.barh(range(len(cols)),per, tick_label = cols)
plt.title('Percentage of null values')
plt.show()
```

```python
## Whenever lot of outliers = replace nan with median or mode
## if numerical feature with nan, create a new column with nan entry = 1 else 0 ,to capture its relation to dependend feature
for col in num_features_with_nan:
    median_value = df[col].median()
    
    df[col + 'nan'] = np.where(df[col].isnull(), 1,0)
    df[col].fillna(median_value, inplace = True)
    
df[num_features_with_nan].isnull().sum()
```

```python
df.head(40)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature] = df['YrSold'] - df[feature]
df.head()
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

# Feature Engineering Part II
## Log normal distribution for skewed features

```python
num_features = ['LotFrontage', 'LotArea','1stFlrSF','GrLivArea']
for f in num_features:
    df[f] = np.log(df[f])
    
```

```python
df.head()
```

## Handling rare categorical feature
We will remove caetgorical variables that are present less than 1% observations


```python
cat_features = df.select_dtypes(include= 'object').columns
cat_features
```

```python
for f in cat_features:
    temp = df.groupby(f)['SalePrice'].count()/len(df)
    temp_df = temp[temp > 0.01].index
    df[f] = np.where(df[f].isin(temp_df), df[f], 'Rare_var')
    
df.head()
```

```python
df.head(100)
```

```python
#Label encoding

for f in cat_features:
    labels_ordered = df.groupby(f)['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i, k in enumerate(labels_ordered, 0)}
    df[f] = df[f].map(labels_ordered)
```

```python
df.head(15)
```

```python
feature_scale = [f for f in df.columns if f not in ['Id', 'SalePrice']]

#Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
#Transform train and test  , and add on the SalePRICE AND Id variable

X_train = pd.concat([df[['Id', 'SalePrice']].reset_index(drop = True),
                    pd.DataFrame(scaler.transform(df[feature_scale]),columns = feature_scale)],
                   axis = 1)
X_train.head()
```
